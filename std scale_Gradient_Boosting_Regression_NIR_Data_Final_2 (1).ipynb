```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import pickle
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
with open('x_train.pkl', 'rb') as file:
    x_train = pickle.load(file)
```

```python
with open('x_test.pkl', 'rb') as file:
    x_test = pickle.load(file)
```

```python
with open('y_train.pkl', 'rb') as file:
    y_train = pickle.load(file)
```

```python
with open('y_test.pkl', 'rb') as file:
    y_test = pickle.load(file)
```

```python
with open('x_train_scaled.pkl', 'rb') as file:
    x_train_scaled = pickle.load(file)
```

```python
with open('x_test_scaled.pkl', 'rb') as file:
    x_test_scaled = pickle.load(file)
```

```python
# Create and train the Gradient Boosting Regressor model
gb_regressor = GradientBoostingRegressor(n_estimators=10, learning_rate=0.1, max_depth=3, random_state=42)
gb_regressor.fit(x_train_scaled, y_train)
```

```python
# Predicting the target values of the test set
y_pred_train = gb_regressor.predict(x_train_scaled)

# RMSE (Root Mean Square Error)
rmse = float(format(np.sqrt(mean_squared_error(y_train, y_pred_train)), '.3f'))
print("\nRMSE: ", rmse)

r2 = r2_score(y_train, y_pred_train)
print(f'R² Score: {r2}')
```

```python
plt.scatter(y_train, y_pred_train, color='blue')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], linestyle='--', color='red', linewidth=2)
plt.xlabel('Actual Values (y_train)')
plt.ylabel('Predicted Values (y_pred_train)')
plt.title('Actual(y_train) vs. Predicted Values(y_pred_train)')
plt.show()
```

```python
# Predicting the target values of the test set
y_pred = gb_regressor.predict(x_test_scaled)

# RMSE (Root Mean Square Error)
rmse = float(format(np.sqrt(mean_squared_error(y_test, y_pred)), '.3f'))
print("\nRMSE: ", rmse)

r2 = r2_score(y_test, y_pred)
print(f'R² Score: {r2}')
```

```python
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', linewidth=2)
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.title('Actual vs. Predicted Values')
plt.show()
```
