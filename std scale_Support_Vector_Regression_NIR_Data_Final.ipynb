```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import pickle
```

```python
with open('x_train.pkl', 'rb') as file:
    x_train = pickle.load(file)
```

```python
with open('x_test.pkl', 'rb') as file:
    x_test = pickle.load(file)
```

```python
with open('y_train.pkl', 'rb') as file:
    y_train = pickle.load(file)
```

```python
# Define the parameter grid for Grid Search (only 'kernel' hyperparameter)
param_grid = {
    'kernel': ['linear', 'rbf']
}
```

```python
# Create an SVR model
svr_model = SVR()
```

```python
# Use GridSearchCV to perform hyperparameter tuning
grid_search = GridSearchCV(svr_model, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
print(f"Best Hyperparameters: {best_params}")
```

```python
# Create and train the SVR model
svr = SVR(kernel='linear', C=0.1, epsilon=0.1, gamma='scale', degree=2)
svr.fit(x_train_scaled, y_train)
```

```python
# Predicting the target values of the test set
y_pred_train = svr.predict(x_train_scaled)

# RMSE (Root Mean Square Error)
rmse = float(format(np.sqrt(mean_squared_error(y_train, y_pred_train)), '.3f'))
print("\nRMSE: ", rmse)

r2 = r2_score(y_train, y_pred_train)
print(f'R² Score: {r2}')
```

```python
plt.scatter(y_train, y_pred_train, color='blue')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], linestyle='--', color='red', linewidth=2)
plt.xlabel('Actual Values (y_train)')
plt.ylabel('Predicted Values (y_pred_train)')
plt.title('Actual(y_train) vs. Predicted Values(y_pred_train)')
plt.show()
```
